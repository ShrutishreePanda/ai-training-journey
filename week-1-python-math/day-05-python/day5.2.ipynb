```python
import pandas as pd
```

```python
df = pd.read_csv("cleaned_titanic.csv")
```

```python
survival_rate = df["Survived"].mean() * 100
print("Overall Survival Rate: {:.2f}%".format(survival_rate))
```

```python
survival_by_class = df.groupby("Pclass")["Survived"].mean() * 100
print(survival_by_class)
```

```python
avg_fare = df.groupby("Pclass")["Fare"].mean()
print(avg_fare)
```

```python
senior_survival = df[df["Age"] > 60]["Survived"].mean() * 100
print("Survival Rate (Age > 60): {:.2f}%".format(senior_survival))
```

```python
df["AgeGroup"] = df["Age"].apply(lambda x: "child" if x < 18 else "adult")
```

```python
agegroup_survival = df.groupby("AgeGroup")["Survived"].mean() * 100
print(agegroup_survival)
```
