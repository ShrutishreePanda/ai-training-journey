```python
import random

def simulate_coin_toss(n=1000):
    outcomes = [random.choice(['H', 'T']) for _ in range(n)]
    heads_prob = outcomes.count('H') / n
    print(f"Probability of Heads: {heads_prob:.2f}")

simulate_coin_toss()
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))
print("Squared:", arr ** 2)

matrix = np.array([[1, 2], [3, 4]])
print("Transpose:\n", matrix.T)
```

```python
import pandas as pd

data = {
    "Name": ["A", "B", "C", "D"],
    "Marks": [85, 90, 78, 92]
}

df = pd.DataFrame(data)

print(df)
print("Average:", df["Marks"].mean())

top_students = df[df["Marks"] > 85]
print(top_students)
```
